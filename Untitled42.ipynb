```python
import os
from pdf2image import convert_from_path

# Ensure output directory exists
output_dir = "docx_pages"
os.makedirs(output_dir, exist_ok=True)

# Convert first 5 pages of PDF to images
pdf_path = "/content/docx.pdf"
images = convert_from_path(pdf_path, dpi=300)

for i, img in enumerate(images[:5]):
    save_path = os.path.join(output_dir, f"page_{i+1}.png")
    img.save(save_path)
    print(f"✅ Saved {save_path}")
```

```python
import random
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
from PIL import Image
```

```python
hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_img(path_to_img, max_dim=512):
    img = Image.open(path_to_img)
    img = img.convert('RGB')
    img.thumbnail((max_dim, max_dim))
    img = np.array(img)
    img = img.astype(np.float32)[np.newaxis, ...] / 255.0
    return tf.constant(img)
```

```python
# Helper: Convert tensor to PIL Image
def tensor_to_image(tensor):
    tensor = tensor * 255
    tensor = tf.cast(tensor, tf.uint8)
    array = tensor.numpy()
    if len(array.shape) > 3:
        array = array[0]
    return Image.fromarray(array)

# Set folder paths
style_dir = "renaissance_dataset"
content_dir = "docx_pages"
output_dir = "stylized_output"
os.makedirs(output_dir, exist_ok=True)

# Select 5 random style images
style_images = random.sample(os.listdir(style_dir), 5)
content_images = sorted([f for f in os.listdir(content_dir) if f.endswith(".png")])[:5]
```

```python
import os
import random

# Define valid image extensions
valid_extensions = [".jpg", ".jpeg", ".png"]

# Filter only valid style images (skip hidden dirs like .ipynb_checkpoints)
style_images = [
    f for f in os.listdir(style_dir)
    if os.path.splitext(f)[1].lower() in valid_extensions and not f.startswith(".") and not os.path.isdir(os.path.join(style_dir, f))
]

# Randomly choose 5 style images
style_images = random.sample(style_images, 5)

# Select first 5 content images (already PNG)
content_images = sorted([f for f in os.listdir(content_dir) if f.endswith(".png")])[:5]

# Apply NST
for i, (content_img_name, style_img_name) in enumerate(zip(content_images, style_images)):
    print(f"Processing Pair {i+1}: {content_img_name} + {style_img_name}")

    content_image = load_img(os.path.join(content_dir, content_img_name))
    style_image = load_img(os.path.join(style_dir, style_img_name))

    stylized_image = hub_model(tf.constant(content_image), tf.constant(style_image))[0]
    final_image = tensor_to_image(stylized_image)
    final_image.save(os.path.join(output_dir, f"stylized_page_{i+1}.png"))
    print(f"✅ Saved stylized_page_{i+1}.png")
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Display original and stylized images side by side
def show_comparison(content_dir, output_dir, num_images=5):
    plt.figure(figsize=(12, num_images * 3))

    for i in range(num_images):
        content_path = os.path.join(content_dir, f"page_{i+1}.png")
        stylized_path = os.path.join(output_dir, f"stylized_page_{i+1}.png")

        content_img = Image.open(content_path)
        stylized_img = Image.open(stylized_path)

        # Show original
        plt.subplot(num_images, 2, 2*i + 1)
        plt.imshow(content_img)
        plt.axis('off')
        plt.title(f"Original Page {i+1}")

        # Show stylized
        plt.subplot(num_images, 2, 2*i + 2)
        plt.imshow(stylized_img)
        plt.axis('off')
        plt.title(f"Stylized Page {i+1}")

    plt.tight_layout()
    plt.show()

# Call it like this:
show_comparison(content_dir="docx_pages", output_dir="stylized_output", num_images=5)
```

```python
from PIL import Image
import numpy as np
import os

# Paths
content_dir = "docx_pages"
stylized_dir = "stylized_output"
blended_dir = "blended_output"
os.makedirs(blended_dir, exist_ok=True)

# Alpha controls the strength of the style
alpha = 0.5

# Get image names (assumes both folders have stylized_page_1.png and page_1.png format)
for i in range(1, 6):
    content_path = os.path.join(content_dir, f"page_{i}.png")
    stylized_path = os.path.join(stylized_dir, f"stylized_page_{i}.png")

    content_img = np.array(Image.open(content_path).convert("RGB"), dtype=np.float32) / 255.0
    stylized_img = np.array(Image.open(stylized_path).convert("RGB"), dtype=np.float32) / 255.0

    # Resize stylized to match content if needed
    if content_img.shape != stylized_img.shape:
        stylized_img = np.array(Image.fromarray((stylized_img * 255).astype(np.uint8)).resize(
            (content_img.shape[1], content_img.shape[0]), Image.BILINEAR)) / 255.0

    # Blend
    blended = (1 - alpha) * content_img + alpha * stylized_img
    blended = np.clip(blended * 255, 0, 255).astype(np.uint8)

    # Save
    Image.fromarray(blended).save(os.path.join(blended_dir, f"blended_page_{i}.png"))
    print(f"✅ Blended saved: blended_page_{i}.png")
```

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Paths
content_dir = "docx_pages"
stylized_dir = "stylized_output"
blended_dir = "blended_output"

# Number of pages to show
num_pages = 5

# Plot
fig, axs = plt.subplots(num_pages, 3, figsize=(12, 3 * num_pages))

for i in range(num_pages):
    content_img = Image.open(os.path.join(content_dir, f"page_{i+1}.png"))
    stylized_img = Image.open(os.path.join(stylized_dir, f"stylized_page_{i+1}.png"))
    blended_img = Image.open(os.path.join(blended_dir, f"blended_page_{i+1}.png"))

    axs[i, 0].imshow(content_img)
    axs[i, 0].set_title("Original")
    axs[i, 0].axis("off")

    axs[i, 1].imshow(stylized_img)
    axs[i, 1].set_title("Stylized")
    axs[i, 1].axis("off")

    axs[i, 2].imshow(blended_img)
    axs[i, 2].set_title("Blended")
    axs[i, 2].axis("off")

plt.tight_layout()
plt.show()
```
